```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

# Extracting 2019's data

```python
url_variable = ['misc','stats', 'shooting','passing']

for variable in url_variable:
    if variable =='misc':
        website_misc_2019 = requests.get(f'https://fbref.com/pt/comps/14/2019/{variable}/2019-Copa-Libertadores-estatisticas')
        html_misc_2019 = website_misc_2019.content
        html_misc_2019 = BeautifulSoup(html_misc_2019)
    elif variable =='stats':
        website_stats_2019 = requests.get(f'https://fbref.com/pt/comps/14/2019/{variable}/2019-Copa-Libertadores-estatisticas')
        html_stats_2019 = website_stats_2019.content
        html_stats_2019 = BeautifulSoup(html_stats_2019)
    elif variable =='shooting':
        website_shoot_2019 = requests.get(f'https://fbref.com/pt/comps/14/2019/{variable}/2019-Copa-Libertadores-estatisticas')
        html_shoot_2019 = website_shoot_2019.content
        html_shoot_2019 = BeautifulSoup(html_shoot_2019)
    else:
        website_pass_2019 = requests.get(f'https://fbref.com/pt/comps/14/2019/{variable}/2019-Copa-Libertadores-estatisticas')
        html_pass_2019 = website_pass_2019.content
        html_pass_2019 = BeautifulSoup(html_pass_2019)
```

```python
#misc
lista_fouls_2019 = []
lista_fouled_2019= []
#passe
lista_passes_2019=[]
#shooting
lista_chuteagol_2019=[]
#stats
lista_posse_2019 = []
lista_gols_2019 = []
lista_pb_2019 = []
lista_pt_2019 = []
lista_crtsa_2019 = []
lista_crtsv_2019 = []
```

```python
sheet_body = html_shoot_2019.find_all('tbody')[0]
```

```python
column_equivalent = ['passes_pct','fouls', 'fouled', 'shots_on_target']

for row in sheet_body.find_all('td', {'data-stat':{'shots_on_target'}}):
    lista_chuteagol_2019 = lista_chuteagol_2019 + list(row)
```

```python
column_equiv = ['possession', 'goals', 'pens_made', 'pens_att', 'cards_yellow', 'cards_red']

cont = 0
for item in column_equiv:
    for row in sheet_body.find_all('td', {'data-stat':f'{item}'}):
        cont = cont + 1
        if cont <= 32:
            lista_posse_2019 = lista_posse_2019 + list(row)
        elif cont <= 64:
            lista_gols_2019 = lista_gols_2019 + list(row)
        elif cont <= 96:
            lista_pb_2019 = lista_pb_2019 + list(row)
        elif cont <= 128:
            lista_pt_2019 = lista_pt_2019 + list(row)
        elif cont <= 160:
            lista_crtsa_2019 = lista_crtsa_2019 + list(row)
        else:
            lista_crtsv_2019 = lista_crtsv_2019 + list(row)
```

```python
find_team = [col_name.text for col_name in html_stats_2019.find_all('th')]
```

```python
team = find_team[35:67]
```

```python
lista_paises=[]
for i in team:
    i=i.upper()
    lista_paises.append(i[:2])
```

```python
lista_times=[]
for i in team:
    lista_times.append(i[3:])
```

```python
ano = [2019]*32
```

```python
first_page_df = pd.DataFrame({'País':lista_paises, 'Equipe':lista_times, 'Posse':lista_posse_2019, 'Gols': lista_gols_2019, 'Penaltis Convertidos': lista_pb_2019, 'Penaltis Marcados':lista_pt_2019, 'Cartão Amarelo':lista_crtsa_2019, 'Cartão Vermelho':lista_crtsv_2019, 'Chutes a Gol':lista_chuteagol_2019, 'Passes Completados':lista_passes_2019, 'Faltas Sofridas':lista_fouls_2019, 'Faltas Causadas':lista_fouled_2019, 'Ano':ano })
```
